```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
X = dataset.iloc[:, [2,3]].values
y= dataset.iloc[:, -1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.25, random_state=0)
```

```python
X_train
```

```python
#featire Scalling
from sklearn.preprocessing import StandardScaler
sc_range= StandardScaler()
X_train=sc_range.fit_transform(X_train)
X_test= sc_range.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state= 0)
classifier.fit(X_train, y_train)  #learn the co-relation between them then predict
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
# Confusion metrix
from sklearn.metrics import confusion_matrix    #function always start lower case
cm =confusion_matrix(y_test, y_pred)
```

```python
cm       #89 correct prediction in rsult and 11 incorrect prediction
```

```python
# Visualize the logistic regression training data
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start=X_set[:,0].min()-1, stop= X_set[:,0].max()+1, step=0.01), #fixed the value of -1 to +1  
                     np.arange(start=X_set[:, 0].min()-1, stop=X_set[:,0].max()+1,  step=0.01)) # for salary and age #step define the bold of line
# countourf use to divided the graph, Listedcolor define the two color #define pixel point to color 0 and 1  
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
            alpha=0.75,cmap= ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())   #limit those salary and age
plt.ylim(X2.min(), X2.max())
for i, j in enumerate (np.unique(y_set)):  #plot all the datapoint
    plt.scatter(X_set[y_set==j,0], X_set[y_set==j, 1], c=ListedColormap(('red', 'green'))(i), label=j)
plt.xlabel("Age")
plt.ylabel('Estimated Salary')
plt.title('Logistic Regression(training set)')
plt.legend()
plt.show()
```
