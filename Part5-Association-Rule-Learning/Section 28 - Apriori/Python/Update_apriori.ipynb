```python
dataset= pd.read_csv("Market_Basket_Optimisation.csv", header= None)
```

```python
transaction =[]
for i in range (0, 7501):  # i define colum and j row [] here define list and convert to string use 
    transaction.append([str(dataset.values[i,j])for j in range(0, 20)])  #divided dataset in list under list
```

```python
#Training apriori algorithm   supoort 3*7/7500
from apyori import apriori  #ita a builking function and import form just c drive
rules= apriori(transaction, min_support=0.003, min_confidence=.2, min_lift=3, min_length=2)  
```
