```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset= pd.read_csv("Mall_Customers.csv")
X= dataset.iloc[:, [3,4]].values        #notice the value otherwise shows error
```

```python
# Import dendogram for identified k values
import scipy.cluster.hierarchy as sch
dendrogram= sch.dendrogram(sch.linkage(X, method='ward'))  #ward use to minimize the variance if not understand then change it and run
plt.title("Dandrogram Diagram")
plt.xlabel('Annual Salary')
plt.ylabel("Eucledean distences or dissimilerity Distence")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc= AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward') #affinity means which method use we use euclidean distance
y_hc =hc.fit_predict(X)
```

```python
# Visualize the Cluster    
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], s=100, c="red", label="Carefull")
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], s=100, c='blue', label="Standard")
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], s=100, c="green", label="Target")
plt.scatter(X[y_hc==3, 0], X[y_hc==3, 1], s=100, c='purple', label="Careless")
plt.scatter(X[y_hc==4, 0], X[y_hc==4, 1], s=100, c="cyan", label="Sensible")
plt.title("K-Means Clusterring")
plt.xlabel("Annual Salary(k$)")
plt.ylabel("Spending in Mall")
plt.legend()
plt.show()
```
