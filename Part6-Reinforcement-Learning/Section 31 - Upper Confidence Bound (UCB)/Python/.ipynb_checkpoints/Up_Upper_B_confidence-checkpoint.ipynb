```python
import math
N=10000
d=10
add_selected=[]
sum_of_reward=[0]*d
num_selection = [0] * d
total_reward=0
for n in range(0, N):
    ad = 0
    maximum_upper_bound=0
    for i in range(0, d):
        if(num_selection[i] > 0):   #selected a add from 1 to 10
            average_reward=sum_of_reward[i]/num_selection[i]
            delta_i= math.sqrt(3/2* math.log(n+1)/num_selection[i])
            upper_bound=average_reward + delta_i
        else:
            upper_bound= 1e400    #pick 0-9 to same add
        if upper_bound > maximum_upper_bound:   # pick up the maximum bound 
            maximum_upper_bound = upper_bound
            ad = i
    add_selected.append(ad)
    num_selection[ad] = num_selection[ad] + 1
    reward= dataset.values[n, ad]
    sum_of_reward[ad]=sum_of_reward[ad] + reward
    total_reward = total_reward + reward
```
